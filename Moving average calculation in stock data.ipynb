```python
import pytse_client as tse
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Download data from TSE Site
data = tse.download(symbols=["کچاد"],include_jdate= True,adjust = False) #adjust = True :قیمت تعدیل یافته سهم
```

```python
df=data['کچاد']
```

```python
type(df),type(data)
```

```python
#save data as CSV if needed
#filename = 'chadormalou.csv'
#df.to_csv(filename)
```

```python
#Plotting trend data
df['open'].plot(label = 'Chadormalou' , figsize = (15,7),legend = True )
```

```python
df.info()
```

```python
df['value'].plot(label = 'Chadormalou' , figsize = (15,7),legend = True )
```

```python
#Moving Average 50 , 200
df['MA50']=df['open'].rolling(50).mean()
df['MA200']=df['open'].rolling(200).mean()
```

```python
df['open'].plot(label = 'Chadormalou' , figsize = (15,7),legend = True )
df['MA50'].plot(label = 'MA50', figsize = (15,7), legend = True)
df['MA200'].plot(label = 'MA200', figsize = (15,7), legend = True )
```

```python
#percentage increase in stock value
df['returns']=(df['close']/df['close'].shift(1))-1
df['returns'].hist(bins = 500,  figsize = (15,7),legend = True)
```
